# Simple polynomial Rregression , that is , 1 independent feature and 1 dependent feature

```python
#import libraries
import numpy as np 
import pandas as pd  
import matplotlib.pyplot as plt 
%matplotlib inline
```

## random way to generate dataset

```python
# quadartic equations , y = 0.5x^2 + 1.5x + 2 + outliers
X = 6* np.random.rand(100,1) - 3
Y = 0.5*X**2 + 1.5*X + 2 + np.random.rand(100,1)

```

```python
plt.scatter(X,Y,color='g')
plt.xlabel('X datset')
plt.ylabel('Y dataset')
plt.show()
```

### train split


```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
## lets implement simple linear regression
from sklearn.linear_model import LinearRegression
regression_1=LinearRegression()
```

```python
regression_1.fit(X_train,Y_train)
```

```python
# performance metrices
from sklearn.metrics import r2_score
score=r2_score(Y_test,regression_1.predict(X_test))
print(score)
```

```python
## lets visualization this model
plt.plot(X_train,regression_1.predict(X_train),color='b')
plt.scatter(X_train,Y_train)
plt.xlabel("X Dataset")
plt.ylabel("Y")
plt.show()
```

### as we can see , since we used linear regression we get good amount of error , lets do this by using polynomial linear Regression

```python
## Plynomial regression
### Lets apply polynomial transformation
from sklearn.preprocessing import PolynomialFeatures 
    
```

```python
poly=PolynomialFeatures(degree=2,include_bias=True)
X_train_poly=poly.fit_transform(X_train)
X_test_poly=poly.transform(X_test)
```

```python
X_train_poly
```

```python
X_test_poly
```

```python
# performance metrices
from sklearn.metrics import r2_score
regression=LinearRegression()
regression.fit(X_train_poly,Y_train)
Y_pred=regression.predict(X_test_poly)
score=r2_score(Y_test,Y_pred)
print(score)
```

```python
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
# best fit line
plt.scatter(X_train,regression.predict(X_train_poly))
plt.scatter(X_train,Y_train)
plt.show()
```

```python
poly=PolynomialFeatures(degree=3,include_bias=True)
X_train_poly=poly.fit_transform(X_train)
X_test_poly=poly.transform(X_test)
```

```python
X_train_poly
```

```python
# performance metrices
from sklearn.metrics import r2_score
regression=LinearRegression()
regression.fit(X_train_poly,Y_train)
Y_pred=regression.predict(X_test_poly)
score=r2_score(Y_test,Y_pred)
print(score)
```

## ## Prediction of new dataset

```python
## Prediction of new dataset
X_new=np.linspace(-3,3,200).reshape(200,1)
X_new_poly=poly.transform(X_new)
X_new_poly
```

```python
Y_new =regression.predict(X_new_poly)
plt.plot(X_new,Y_new,"r-",linewidth=2,label=" New Predictions")
plt.plot(X_train,Y_train,"b.",label="Training Points")
plt.plot(X_test,Y_test,"g.",linewidth=2,label="Testing Points")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

## Pipelining Concepts

```python
from sklearn.pipeline import Pipeline
```

```python
def poly_regression(degree):
    X_new=np.linspace(-3,3,200).reshape(200,1)
    
    poly_features=PolynomialFeatures(degree=degree,include_bias=True)
    lin_reg=LinearRegression()
    poly_regression=Pipeline([
        ("PolynomialFeatures",poly_features),
        ("lin_reg",lin_reg)
    ])
    poly_regression.fit(X_train,Y_train)
    Y_pred_new=poly_regression.predict(X_new)
    # plotting prediction line 
    plt.plot(X_new,Y_pred_new,'r',label="Degree " +str(degree),linewidth=3)
    plt.plot(X_train,Y_train,"b.",linewidth=3)
    plt.plot(X_test,Y_test,"g.",linewidth=3)
    plt.legend(loc='upper left')
    plt.xlabel("X")
    plt.ylabel("y")
    plt.axis([-4,4,0,10])
    plt.show()
    
```

```python
poly_regression(15)
```
