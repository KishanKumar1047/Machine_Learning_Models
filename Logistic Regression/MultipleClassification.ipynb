## Logistic Regression Implementation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns      
%matplotlib inline
```

```python
## import datasets from sklearn
from sklearn.datasets import make_classification
# make_classification is a utility function to generate a synthetic classification dataset.
# It's commonly used for testing and demonstrating machine learning models.
```

```python
#create dataset
X,y=make_classification(n_samples=1000,n_features=10,n_informative=3,n_classes=3,random_state=30)
```

```python
X
```

```python
#conerting x into dataframe
pd.DataFrame(X)
```

```python
y
```

### Train-Test Split

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=42)
```

### Implement Logistic Regression

```python
## Model Selection
from sklearn.linear_model import LogisticRegression
logistic=LogisticRegression(multi_class='ovr')
```

```python
logistic.fit(X_train,y_train)
```

```python
y_pred=logistic.predict(X_test)
print(y_pred)
```

```python
# logistic.predict_proba(X_test)
```

### Performance metrices

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report 
```

```python
score=accuracy_score(y_test,y_pred)
print(score)
cm=confusion_matrix(y_test,y_pred)
print(cm)
print(classification_report(y_test,y_pred))
```

# Hperparameter and Cross Validation

# Gridsearch CV

```python
model=LogisticRegression()
penalty=['l1','l2','elasticnet']
c_values=[100,10,1.0,0.1,0.01]
solver=['newton-cg','lbgfs','liblinear','sag','saga']
```

```python
#key-value pair
params=dict(penalty=penalty,C=c_values,solver=solver)
```

```python
#stratisfied kfold
from sklearn.model_selection import StratifiedKFold
cv=StratifiedKFold()
```

```python
from sklearn.model_selection import GridSearchCV
grid=GridSearchCV(estimator=model,param_grid=params,scoring='accuracy',cv=cv,n_jobs=-1)
```

```python
print(grid)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_pred=grid.predict(X_test)
```

```python
score=accuracy_score(y_test,y_pred)
print(score)
cm=confusion_matrix(y_test,y_pred)
print(cm)
print(classification_report(y_test,y_pred))
```

# Randomized searchCV

```python
model=LogisticRegression()
```

```python
from sklearn.model_selection import RandomizedSearchCV
randomCV=RandomizedSearchCV(estimator=model,param_distributions=params,scoring='accuracy',cv=5)
```

```python
print(randomCV.fit(X_train,y_train))
```

```python
randomCV.best_score_
```

```python
randomCV.best_params_
```

```python
y_pred=randomCV.predict(X_test)
```

```python
score=accuracy_score(y_test,y_pred)
print(score)
cm=confusion_matrix(y_test,y_pred)
print(cm)
print(classification_report(y_test,y_pred))
```
